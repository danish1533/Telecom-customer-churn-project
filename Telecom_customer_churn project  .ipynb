```python
X=df.drop('Churn',axis=1)
y=df['Churn']
```

# balancing 

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample=SMOTE()
X,y=oversample.fit_resample(X,y)
```

# now standard scaling 

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_sc=sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

# tring with dtc 

```python
maxacc=0
maxrs=0
for i in range(1,250):
    X_train,X_test,y_train,y_test=train_test_split(X_sc,y,random_state=i,test_size=0.2)
    dtc=DecisionTreeClassifier()
    dtc.fit(X_train,y_train)
    y_pred=dtc.predict(X_test)
    acc=accuracy_score(y_test,y_pred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print('the best accuracy score in dtc ',maxacc,'on random state is ',maxrs)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_sc,y,test_size=0.2,random_state=114)
dtc.fit(X_train,y_train)
y_pred=dtc.predict(X_test)
print(classification_report(y_test,y_pred))
```

# now training the data by logistic 

```python
maxacc=0
maxrs=0
for i in range(1,250):
    X_train,X_test,y_train,y_test=train_test_split(X_sc,y,random_state=i,test_size=0.2)
    lr=LogisticRegression()
    lr.fit(X_train,y_train)
    y_pred=lr.predict(X_test)
    acc=accuracy_score(y_test,y_pred)
    if acc>maxacc:
        maxacc=acc
        maxrs=i
print('the best acuracy score is ',maxacc,'on random state ',maxrs)
    
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_sc,y,test_size=0.2,random_state=134)
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
print(classification_report(y_test,y_pred))
```

# now deciding the parameters  by gridsearchcv

```python
param_grid={'criterion':['gini','entropy'],'max_depth':[i for i in range(1,20,3)],'min_samples_leaf':[i for i in range(1,20,3)],'min_samples_split':[ i for i in range(1,20)]}
grid_search=GridSearchCV(dtc,param_grid,cv=5)
grid_search.fit(X_train,y_train)
print('best param',grid_search.best_params_)
```

```python
final_model=DecisionTreeClassifier(criterion='gini', max_depth= 10, min_samples_leaf= 1, min_samples_split= 3)
final_model.fit(X_train,y_train)
y_pred=final_model.predict(X_test)
print('accuracy_score\n',accuracy_score(y_test,y_pred))
```
